```python
import pandas as pd
df = pd.read_csv("Train.csv")
```

```python
# Filter the rows where 'YearMade' is >= 2000
df = df[df['YearMade'] >= 1995]
```

```python
missing_data = df.isna().sum()
print(missing_data[missing_data>0])
```

```python
missing_percentages = (df.isna().sum() / len(df)) * 100
missing_percentages_sorted=missing_percentages[missing_percentages>0].sort_values(ascending=False)
print(missing_percentages_sorted)

```

```python
col_to_drop=missing_percentages_sorted[missing_percentages_sorted>90].index
df=df.drop(columns=col_to_drop)
print(col_to_drop)
```
